# Ch3.4 Classifying movie reviews: a binary classification example
This example comes from chapter 3.4 in the book (page 68). Here we learn how to classify movie reviews from IMDB as being positive or negative based on the text content of the review.

The IMDB dataset contains 50,000 reviews, which have been split into two sets (25,000 each) for trainihg and testing. Half the examples are positive reviews and the other half are negative.

## Load the IMDB dataset
The IMDB dataset is already included with Keras so we can import directly from there.

```python
from keras.datasets import imdb
(train_d,train_l), (test_d,test_l) = imdb.load_data(num_words=10000)
```

The data is loaded into the training and test sets. Each contains the actual data (list of words, i.e. train_d) and the labels (train_l) indicating whether the review was positive or negative. The argument num_words allows us to specify the number of words to keep. The program will keep the specified number of most frequently used words in the dataset, in this case 10,000. This means that rare words will be disgarded.

## Looking at the data
We can check the size of each of these NumPy arrays using the .size method:

```python
import numpy as np
train_d.size, train_l.size, test_d.size, test_l.size
```

Let's now take a look at the data. Note that here each review is a list of word indices (encoding a sequence of words). This means that each word has a unique index that we can look up to convert the numbers back to text. The list of numbers for the first review in the training dataset is printed below.

```python
train_d[0]
```

The label is either a 0 or 1 (for negative and positive). For the first example in the training set:

```python
train_l[0]
```

When we loaded the data, we choose to only keep the 10,000 most commonly used words in the dataset. We can check that no word index exceeds this value using the code below:

```python
max([max(seq) for seq in train_d])
```

A dictionary is used to map words to the integer indexes. We can see this dictionary by using the get_word_index() method on the IMDB data. Note below that this includes all of the words in the data so the index exceeds 10,000 here.

```python
imdb.get_word_index()
```

We can use the dictionary to decode the review from integers to words. The current dictionary (above) maps words to integers. To decode the reviews, we need to reverse this so that the dictionary maps integers to words. Remember that in the dictinary above, the items in quotes to the left are the key, and on the right is the value associated with that key.

```python
# This current dictionary maps words to integers
word_index = imdb.get_word_index()
# Here we want to create a new dictionary which does the reverse (maps integers to words)
rev_word_index = dict([(value,key) for (key,value) in word_index.items()])
```

```python
rev_word_index
```

Now we can use this reversed dictionary to decode the reviews. Here we need to use the 'get' method on the dictionary to find the word corresponding to each number. If the number is not included in our dictionary (i.e. greater than 10,000) then these are replaced with a question mark. Note here that the first three indeces (0,1,2) are reserved (for padding, start of sequence and unknown respectively), so we need to offset our index by 3 to return the correct word.

```python
decoded_review = ' '.join([rev_word_index.get(i-3, '?') for i in train_d[0]])
```

```python
decoded_review
```

# Data preparation
Neural networks work with tensors, so we cannot feed lists of integers into a neural network. There are two options to turn lists into tensors. The first is to pad the lists so that each have the same length and then use an embedding layer in the neural network. For this example we will use the second option, which is to turn the lists into vectors of 0s and 1s. This is known as one-hot encoding. 

## One-hot encoding
This means that we turn our lists into vectors of 0's and 1's, where a 1 would indicate that the word is present in the review. In this example, we effectively turn each review into a 10,000 dimensional vector (since we selected to include the 10,000 most frequently used words). This generally leads to a very sparse vector/matrix (since most of the values will be zero).

Here we will define a function and use NumPy to do this manually. In the code below we use the enumerate() method to create a counter for our iterator (see example here https://www.geeksforgeeks.org/enumerate-in-python/). Basically this outputs i, which is the row or training example number.

```python
import numpy as np

def vectorize_sequences(sequences,dimension=10000):
    result = np.zeros((len(sequences),dimension))
    for i,sequence in enumerate(sequences):
        result[i,sequence] = 1
    return result

# Vectorize the training and test data and labels
x_train = vectorize_sequences(train_d)
x_test = vectorize_sequences(test_d)  
```

```python
x_train
```

We should also vectorize the labels:

```python
# Our vectorized labels
y_train = np.asarray(train_l).astype('float32')
y_test = np.asarray(test_l).astype('float32')
```

# Building a neural network

Here our input data is vectors with labels that are scalars (i.e. 0 or 1). A simple stack fo fully connected layers with ReLU activations works well in these cases. For a stack fo dense layers, we have two key architecture decisions to make:
- How many layers to use
- How many hidden units for each layer

The more layers and hidden units that we use, the more freedom the network has to learn complex representations of the input data. However, we need to be careful that we don't overfit on the training data and build amodel that does not generalize well to data it has not seen before. The principals for making these decisions are covered in chapter 4. For now we will just trust that the following are good architecture choices:
- Two intermediate layers, each with 16 units
- A third layer that wil output the scalar prediction regarding the sentiment of the current review

Here we will use relu as the activation function for the intermediate layers and a sigmoid to output the probability (i.e. score of 1 indicates positive).

## Define the model

```python
from keras import models
from keras import layers
```

```python
model = models.Sequential()
model.add(layers.Dense(16,activation='relu',input_shape=(10000,)))
model.add(layers.Dense(16,activation='relu'))
model.add(layers.Dense(1,activation='sigmoid'))
```

## Compile the model
Next we have to select the loss function (i.e. a score that tells the network how well it is doing), an optimizer (which tells the network how the parameters/weights should be updated based on the gradient of the loss function) and metric to measure the accuracy of the model on training and test data. 

### Loss function
Because we are dealing with a binary classification problem and the output is a probability, the best choice is the binary cross-entropy loss (also known as log loss). Some more information can be found here: https://ml-cheatsheet.readthedocs.io/en/latest/loss_functions.html . It is not the only choice, but typically works best when we have models that output probabilities. 

### Optimizer
For now we will use the RMS prop optimizer, which is generally a good enough choice whatever the problem is. 

```python
model.compile(optimizer='rmsprop',
             loss='binary_crossentropy',
             metrics=['accuracy'])
```

Because the rmsprop, binary_crossentropy and accuracy are packaged inside Keras, we can pass them as strings as above. However, if we want to change the parameters of the optimizer for instance then we need to pass an optimizer class for the optimizer argument:

```python
#from keras import optimizers

#model.compile(optimizer=optimizers.RMSprop(lr=0.001),
             #loss='binary_crossentropy',
             #metrics=['accuracy'])
```

# Validation
To allow us to monitor the model during training on data it has not seen before we will create a validation set. Here we will use 10,000 samples for the validation (leaving 15,000 for the training). 

```python
train_l[0:10]
```

Since the data appear to be sorted randomly (see above), we can just assign the first 10,000 reviews for the validation set and the remainder for the training.

```python
x_val = x_train[:10000]
partial_x_train = x_train[10000:]
y_val = y_train[:10000]
partial_y_train = y_train[10000:]
```

# Train model
Now we will train the model for 20 epochs. One epoch means one iteration over all samples in the training dataset. We can't run all 15,000 examples at once, so we use mini-batches until the model has seen each training example (meaning one epoch). In this case we will use min-batches of 512 samples. 

While we are training the model, we can measure the performance of the model on the validation set using the validation_data argument.

```python
model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['acc'])
```

```python
history = model.fit(partial_x_train,
                   partial_y_train,
                   batch_size=512,
                   epochs=20,
                   validation_data=(x_val,y_val))
```

# Plotting loss and accuracy
The call to model.fit() returns a History object. This object has a member called history which is a dictionary that records everything that happenened during training.

```python
history_dict = history.history
```

```python
history_dict.keys()
```

# Use matplotlib to plot data
Here we import matplotlib. Note we need to use the Jupyter magic to allow us to plot within the Jupyter notebook (I think this is now by default).


```python
import matplotlib.pyplot as plt

loss_values = history_dict['loss']
val_loss_values = history_dict['val_loss']
epochs = range(1,21)

plt.plot(epochs,loss_values,'bo',label='Training loss')
plt.plot(epochs,val_loss_values,'b',label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

As expected, the training loss decreases over time as the model adjusts to the training data. However, we can also see that from around the third epoch, the validation loss starts to increase. This indicates that we are over-fitting on the training data. In this case, we would want to keep the model after epoch three.

```python
acc_values = history_dict['acc']
val_acc_values = history_dict['val_acc']
epochs = range(1,21)

plt.plot(epochs,acc_values,'bo',label='Training accuracy')
plt.plot(epochs,val_acc_values,'b',label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

We see a similar trend with the accuracy. The model gets more accurate on the training data with increasing number of epochs, reaching close to 100% accuracy. However, for the validation data, we see improving results until around epoch three or four, after which the accuracy gets worse. Again, this indicates overfitting.

# Retrain model from scratch
Now we will just train the model for four epochs (since beyond that it was overfitting) and then use this to see how well it makes predictions on the test dataset.

```python
model = models.Sequential()
model.add(layers.Dense(16,activation='relu',input_shape=(10000,)))
model.add(layers.Dense(16,activation='relu'))
model.add(layers.Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='rmsprop',
             loss='binary_crossentropy',
             metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,batch_size=512,epochs=4)
```

```python
results = model.evaluate(x_test,y_test)
```

```python
results
```

With this fairly naive approach, we have been able to achieve around 87% accuracy on the test data. We can do much better with state-of-the-art approaches.

# Making predictions
Now that we have built a modelm we can make predictions (i.e. the likelihood of reviews being positive or negative). We do this with the predict method.

```python
model.predict(x_test)
```
